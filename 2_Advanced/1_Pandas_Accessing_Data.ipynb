```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

# Lệnh .iloc[]
- Để cho ra thông tin của một row chỉ định
- Row đầu tiên sẽ có index là 0


```python
df.iloc[0]
```

```python
df.iloc[0][15]
```

```python
df.iloc[0, 15]
```

# Cách dùng iloc để lấy thông tin những ô cần thiết
- Lấy từ hàng 3 tới hàng 4 
- Và từ cột 16 tới cột 17

```python
df.iloc[[2,3],[15,16]]
```

```python
# Lấy các hàng từ 0 đến 8 (tổng cộng 9 hàng)
df.iloc[:9]
```

```python
df.iloc[:, :5]
# Lấy tất cả các hàng và các cột từ đầu đến cột thứ 5 (không bao gồm cột thứ 5)
```

```python
df.loc[:, 'job_title_short']
```

```python
df['job_title_short']
```
