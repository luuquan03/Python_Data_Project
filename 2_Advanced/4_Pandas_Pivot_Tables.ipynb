```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Tạo 1 pitvot table với country là hàng, title là cột để so sánh lương median
pivot_df = df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
```

```python
# Bộ lọc 6 quốc gia có nhiều jobs nhất
top_countries = df['job_country'].value_counts().head(6).index
```

```python
# Lọc các title mà mình quan tâm
job_titles = ['Data Scientist', 'Data Engineer', 'Data Analyst']
```

```python
# Lọc pivot table để chỉ giữ lại các quốc gia và job title mình quan tâm
pivot_df = pivot_df.loc[top_countries, job_titles]

# Vẽ biểu đồ cột
pivot_df.plot(kind='bar')
plt.title('Median Salary by Job Title and Country')
plt.ylabel('Median Yearly Salary (USD)')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Job Title')
plt.show()
```
