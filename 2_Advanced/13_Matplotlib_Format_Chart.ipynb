```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
fig, ax = plt.subplots(1, 2)

df['job_title_short'].value_counts().plot(kind='bar', ax=ax[1])
df['job_schedule_type'].value_counts().head(3).plot(kind='bar', ax=ax[0])

fig.tight_layout()
```

# Example - Counts of Top Skills in Job Postings

```python
df_skills = df.copy()
df_skills = df.explode('job_skills')
skills_count = df_skills.groupby(['job_title_short', 'job_skills']).size()
df_skills_count = skills_count.reset_index(name='skill_count').sort_values(by='skill_count', ascending=False)

df_skills_count
```

```python
job_title = ['Data Scientist', 'Data Engineer', 'Data Analyst']
```

```python
fig, ax = plt.subplots(3, 1)

for i, job_title in enumerate(job_title):
    df_skills_count[df_skills_count['job_title_short'] == job_title].head(5).plot(kind='barh', x='job_skills', y='skill_count', ax=ax[i], title=job_title, legend=False)
    ax[i].invert_yaxis()
    ax[i].set_xlabel('')

fig.tight_layout()
fig.suptitle('Top 5 Skills by Job Title', y=1.02)
plt.show()
```
